```python
def G_ref(x1, T):
    R=8.3145
    x2=1-x1
    G=x1*R*T*np.log(x1)+x2*R*T*np.log(x2)
    return G
```

```python
def G_cmb(x1, T, r1, r2, q1, q2, z):
    R=8.3145
    x2=1-x1
    phi1=r1*x1/(r1*x1+r2*x2)
    phi2=r2*x2/(r1*x1+r2*x2)
    theta1=q1*x1/(q1*x1+q2*x2)
    theta2=q2*x2/(q1*x1+q2*x2)
    G=R*T*(z*0.5*(x1*q1*np.log(theta1/phi1)+x2*q2*np.log(theta2/phi2)))#(x1*np.log(phi1/x1)+x2*np.log(phi2/x2)))+
    return G
```

```python
def G_res(x1, T, q1, q2, w12, w21):
    R=8.3145
    x2=1-x1
    theta1=q1*x1/(q1*x1+q2*x2)
    theta2=q2*x2/(q1*x1+q2*x2)
    tau12=np.exp(-w12/T)
    tau21=np.exp(-w21/T)
    rho1=theta1+theta2*tau21
    rho2=theta1*tau12+theta2
    G=-R*T*(x1*q1*np.log(rho1)+x2*q2*np.log(rho2))
    return G
```

## Activity coefficient

```python
x1=sp.Symbol('x1')
x2=sp.Symbol('x2')
def G_u(x1, x2, T, r1, r2, q1, q2, z, w12, w21):
    R=8.31451
    Gref=x1*R*T*sp.ln(x1)+x2*R*T*sp.ln(x2)
    phi1=r1*x1/(r1*x1+r2*x2)
    phi2=r2*x2/(r1*x1+r2*x2)
    theta1=q1*x1/(q1*x1+q2*x2)
    theta2=q2*x2/(q1*x1+q2*x2)
    Gcmb=R*T*((x1*sp.ln(phi1/x1)+x2*sp.ln(phi2/x2))+z*0.5*(x1*q1*sp.ln(theta1/phi1)+x2*q2*sp.ln(theta2/phi2)))
    tau12=sp.exp(-w12/T)
    tau21=sp.exp(-w21/T)
    rho1=theta1+theta2*tau21
    rho2=theta1*tau12+theta2
    Gres=-R*T*(x1*q1*sp.ln(rho1)+x2*q2*sp.ln(rho2))
    G=Gref+Gcmb+Gres
    return G
```

```python
def gamma(xa, xb, T):
    mu1=G_u(x1, x2, T, rA, rB, qA, qB, z, wAB, wBA)+G_u(x1, x2, T, rA, rB, qA, qB, z, wAB, wBA).diff(x1)-(x2)*G_u(x1, x2, T, rA, rB, qA, qB, z, wAB, wBA).diff(x2)-(x1)*G_u(x1, x2, T, rA, rB, qA, qB, z, wAB, wBA).diff(x1)
    mu2=G_u(x1, x2, T, rA, rB, qA, qB, z, wAB, wBA)+G_u(x1, x2, T, rA, rB, qA, qB, z, wAB, wBA).diff(x2)-(x2)*G_u(x1, x2, T, rA, rB, qA, qB, z, wAB, wBA).diff(x2)-(x1)*G_u(x1, x2, T, rA, rB, qA, qB, z, wAB, wBA).diff(x1)
    gamma1=sp.exp((mu1.subs(x1, xa).subs(x2, xb))/R/T)/xa
    gamma2=sp.exp((mu2.subs(x1, xa).subs(x2, xb))/R/T)/xb
    return gamma1, gamma2
```
